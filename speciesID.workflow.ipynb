### species ID using kmcp:

1. download reference sequences
2. build index
3. analyse sample
4. find top match (also try using kmcp profile function)

```python
#1
#mamba install -c conda-forge ncbi-datasets-cli
datasets download genome taxon "Metamycoplasma" --assembly-source RefSeq --assembly-level complete --include genome #--reference
#unzip ncbi_dataset.zip

#2
#mamba install -c bioconda kmcp
kmcp compute -k 21 -n 10 -l 150 -I ncbi_dataset/data/ -O kmcp.myco
kmcp index -f 0.3 -n 1 -I kmcp.myco/ -O kmcp.myco.DB

#3
kmcp search -d kmcp.myco.DB/ -o test.tsv.gz -1 fastq/SRR11901822_1.fastq.gz -2 fastq/SRR11901822_2.fastq.gz
```

```python
#4

import pandas as pd

kmcp_results= pd.read_csv("test.tsv.gz",sep="\t")
kmcp_results
```

```python
kmcp_results.groupby(["target"]).agg({'qCov':['sum']}).reset_index().sort_values(by=[('qCov', 'sum')], ascending=False)
```
